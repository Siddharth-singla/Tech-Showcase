```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


```

```python
df = pd.read_csv('EDA_Housing.csv')
df
```

```python
df['bedrooms']
```

```python
sns.histplot(df['area'],color='g',bins=30,kde=True,)
plt.title('housing bar graph')
plt.yticks(np.arange(0,81, step=100))
plt.xticks(np.arange(0,5, step=1))
plt.grid()
plt.show()
```

```python
df.corr()
```

```python
sns.heatmap(df.corr() ,annot=True)
plt.title('housing heatmap')
plt.show()

```

```python
sns.boxplot(df['stories'])
plt.title('housing boxplot')
plt.figure(figsize=(100,40))
plt.show()
```

```python
sns.distplot(df['parking'])
plt.show()
```
